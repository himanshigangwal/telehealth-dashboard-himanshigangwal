## Data Simulation

```python
import pandas as pd
import numpy as np

np.random.seed(42)

days = pd.date_range(start="2026-04-01", periods=30)

patients = ["P1", "P2", "P3"]

data = []

for patient in patients:
    heart_rate = np.random.randint(60, 130, size=30)
    steps = np.random.randint(1000, 10000, size=30)
    pain = np.random.randint(1, 10, size=30)
    
    for i in range(30):
        data.append([patient, days[i], heart_rate[i], steps[i], pain[i]])

df = pd.DataFrame(data, columns=["Patient", "Date", "HeartRate", "Steps", "PainScore"])

df.head()
```

## Data Simulation

Synthetic patient data was generated for 30 days including:
- Heart Rate
- Steps Walked
- Pain Score

This simulates real-world remote monitoring using wearable devices and IoT systems.

## Dashboard – Line Graphs

### Heart Rate Trend

```python
import matplotlib.pyplot as plt

for patient in df["Patient"].unique():
    subset = df[df["Patient"] == patient]
    plt.plot(subset["Date"], subset["HeartRate"], label=patient)

plt.legend()
plt.title("Heart Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Heart Rate")
plt.xticks(rotation=45)
plt.show()
```

### Steps Walked

```python
for patient in df["Patient"].unique():
    subset = df[df["Patient"] == patient]
    plt.plot(subset["Date"], subset["Steps"], label=patient)

plt.legend()
plt.title("Steps Walked Over Time")
plt.xlabel("Date")
plt.ylabel("Steps")
plt.xticks(rotation=45)
plt.show()
```

### Pain Score

```python
for patient in df["Patient"].unique():
    subset = df[df["Patient"] == patient]
    plt.plot(subset["Date"], subset["PainScore"], label=patient)

plt.legend()
plt.title("Pain Score Trend")
plt.xlabel("Date")
plt.ylabel("Pain Score")
plt.xticks(rotation=45)
plt.show()
```

## Dashboard Visualization

### Heart Rate Trend
Shows variation in heart rate over time for multiple patients.

### Steps Walked
Represents daily activity levels of patients.

### Pain Score Trend
Helps track recovery progress and discomfort levels.

These visualizations help physiotherapists monitor patients remotely.

## Alerts System

```python
alerts = df[df["HeartRate"] > 120]

print(" High Risk Patients:")
print(alerts[["Patient", "Date", "HeartRate"]])
```

## Alerts and Insights

Patients with heart rate greater than 120 bpm are flagged as high risk.

### Insights
- Sudden spikes in heart rate may indicate health issues.
- Low step count combined with high pain score may indicate poor recovery.
- Continuous monitoring helps early intervention.

This system improves patient safety and decision-making.

### Risk Summary

```python
risk_summary = alerts["Patient"].value_counts()

print("\nPatients at Risk:")
print(risk_summary)
```
